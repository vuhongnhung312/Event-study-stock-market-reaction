```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile(r'/Users/hongnhungvu/Desktop/Thesis_Data/CDNB_matched.xlsx')
df_NB = pd.read_excel(xls)
df_NB
```

```python
df_NB = df_NB[(df_NB['RegisterDate'] < df_NB['ResultReportDate']) &
               (df_NB['DateStartExpected'] < df_NB['DateEndExpected']) &
               (df_NB['DateActionFrom'] <= df_NB['DateActionTo'])]

df_NB = df_NB.drop(df_NB[df_NB.RegisterBuyVolume == 0].index)
df_NB = df_NB.drop(df_NB[df_NB.RegisterSellVolume == 0].index)
df_NB = df_NB.drop(df_NB[df_NB.VolumeAfter < 0].index)

df_NB = df_NB[(df_NB['RegisterVolumeBefore'] == df_NB['VolumeBefore'])]

df_NB = df_NB.dropna(subset=['RegisterBuyVolume','RegisterSellVolume'],how='all')
df_NB = df_NB[df_NB[['RegisterBuyVolume','RegisterSellVolume']].isna().any(axis=1)]
df_NB = df_NB.dropna(subset=['BuyVolume','SellVolume'],how='all')
df_NB = df_NB[df_NB[['BuyVolume','SellVolume']].isna().any(axis=1)]

df_NB['VolumeAfterCheck'] = df_NB.VolumeBefore - df_NB.SellVolume.fillna(0) + df_NB.BuyVolume.fillna(0)
df_NB['RegisterVolumeAfterCheck'] = df_NB.RegisterVolumeBefore - df_NB.RegisterSellVolume.fillna(0) + df_NB.RegisterBuyVolume.fillna(0)

df_NB
```

```python
df_NB['VolumeDiffCheck'] = df_NB['VolumeAfterCheck'] - df_NB['VolumeAfter']
df_NB = df_NB.drop(df_NB[df_NB.VolumeDiffCheck != 0].index)

df_NB['RegisterVolumeAfterDiffCheck'] = df_NB['RegisterVolumeAfterCheck'] - df_NB['RegisterVolumeAfter']
df_NB = df_NB.drop(df_NB[df_NB.RegisterVolumeAfterDiffCheck != 0].index)

df_NB = df_NB.drop(['VolumeAfterCheck','VolumeDiffCheck','RegisterVolumeAfterCheck','RegisterVolumeAfterDiffCheck'],axis=1)

df_NB
```

```python
df_NB.to_excel(r'/Users/hongnhungvu/Desktop/Thesis_Data/CDNB_cleaned.xlsx',index=False,header=True)
```
