```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile(r'/Users/hongnhungvu/Desktop/Thesis_Data/NLQ_matched.xlsx')
df_NLQ = pd.read_excel(xls)
df_NLQ
```

```python
df_NLQ = df_NLQ[(df_NLQ['RegisterDate'] < df_NLQ['ResultReportDate']) &
               (df_NLQ['DateStartExpected'] < df_NLQ['DateEndExpected']) &
               (df_NLQ['DateActionFrom'] <= df_NLQ['DateActionTo'])]

df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.RegisterBuyVolume == 0].index)
df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.RegisterSellVolume == 0].index)
df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.RegisterVolumeAfter < 0].index)
df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.VolumeAfter < 0].index)

df_NLQ = df_NLQ[(df_NLQ['RegisterVolumeBefore'] == df_NLQ['VolumeBefore'])]

df_NLQ = df_NLQ.dropna(subset=['RegisterBuyVolume','RegisterSellVolume'],how='all')
df_NLQ = df_NLQ[df_NLQ[['RegisterBuyVolume','RegisterSellVolume']].isna().any(axis=1)]
df_NLQ = df_NLQ.dropna(subset=['BuyVolume','SellVolume'],how='all')
df_NLQ = df_NLQ[df_NLQ[['BuyVolume','SellVolume']].isna().any(axis=1)]

df_NLQ['VolumeAfterCheck'] = df_NLQ.VolumeBefore - df_NLQ.SellVolume.fillna(0) + df_NLQ.BuyVolume.fillna(0)
df_NLQ['RegisterVolumeAfterCheck'] = df_NLQ.RegisterVolumeBefore - df_NLQ.RegisterSellVolume.fillna(0) + df_NLQ.RegisterBuyVolume.fillna(0)

df_NLQ
```

```python
df_NLQ['VolumeDiffCheck'] = df_NLQ['VolumeAfterCheck'] - df_NLQ['VolumeAfter']
df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.VolumeDiffCheck != 0].index)

df_NLQ['RegisterVolumeAfterDiffCheck'] = df_NLQ['RegisterVolumeAfterCheck'] - df_NLQ['RegisterVolumeAfter']
df_NLQ = df_NLQ.drop(df_NLQ[df_NLQ.RegisterVolumeAfterDiffCheck != 0].index)

df_NLQ = df_NLQ.drop(['VolumeAfterCheck','VolumeDiffCheck','RegisterVolumeAfterCheck','RegisterVolumeAfterDiffCheck'],axis=1)

df_NLQ
```

```python
df_NLQ.to_excel(r'/Users/hongnhungvu/Desktop/Thesis_Data/NLQ_cleaned.xlsx',index=False,header=True)
```
