Step 1: Import relevant packages

```python
import numpy as np
import pandas as pd
```

```python
xls = pd.ExcelFile(r'/Users/hongnhungvu/Desktop/Thesis_Data/Data_IT/Insider_Trading_events.xlsx')
df = pd.read_excel(xls)
```

Step 2: Remove accents

```python
df.rename(columns={"DateBuyExpected":"DateStartExpected","DateSellExpected":"DateEndExpected"},inplace=True)

import unidecode
def remove_accent(data):
    list_col = ['Title_Event','TypeName','DTTHLQ','StatusName','PositionCD','DTTHLQ','DTTHCD','DTLQLQ','NVTH','ExtraPositionNLQ','ExtraPositionNLQEx','RelationShipType']
    
    for colname in list_col:
        data[colname]=data[colname].apply(str)
        data[colname]=data[colname].apply(unidecode.unidecode)
    return data

remove_accent(df)
```

Step 3: Clean data before matching

```python
#Only keep all Report Date before 2021
df = df[(df['ReportDate']<'2021-01-01')]
```

```python
#Match events of affiliated persons. df1 is the dataframe contains only events of affiliated persons
df1 = df[df['TypeName']=='GD cua nguoi lien quan']

#Get register events first, rename ReportDate as RegisteredDate, create df1_dangky & df1_ketqua 
df1_dangky = df1[df1['StatusName']=='Dang ky']
df1_dangky.rename(columns={"ReportDate":"RegisterDate"}, inplace=True)
df1_dangky = df1_dangky[['StockCode','RegisterDate','DTLQLQ','DTTHLQ','DateStartExpected','DateEndExpected']]


df1_ketqua = df1[df1['StatusName']=='Ket qua']
df1_ketqua.rename(columns={"ReportDate":"ResultReportDate"}, inplace=True)
df1_dangky
```

```python
df1_ketqua
```

```python
#Match df1_ketqua with df1_dangky using 'StockCode','DTLQLQ','DTTHLQ','DateStartExpected','DateEndExpected'
df1_ketqua = df1_ketqua.merge(df1_dangky, left_on=['StockCode','DTLQLQ','DTTHLQ','DateStartExpected','DateEndExpected'], right_on=['StockCode','DTLQLQ','DTTHLQ','DateStartExpected','DateEndExpected'])
df1_ketqua
```

```python
df1_ketqua = df1_ketqua.reindex(columns=['StockCode','RegisterDate','ResultReportDate','DateStartExpected','DateEndExpected','DateActionFrom','DateActionTo','TypeName','RegisterVolumeBefore','RegisterVolumeBeforePercent','RegisterBuyVolume','RegisterSellVolume','RegisterVolumeAfter','RegisterVolumeAfterPercent','VolumeBefore','VolumeBeforePercent','BuyVolume','SellVolume','VolumeAfter','VolumeAfterPercent','Title_Event','PositionCD','DTTHCD','DTTHLQ','DTLQLQ','NVTH','ExtraPositionNLQ','ExtraPositionNLQEx','RelationShipType'])
df1_ketqua = df1_ketqua.dropna(subset=['RegisterDate','ResultReportDate','DateStartExpected','DateEndExpected','DateActionFrom','DateActionTo'])
df1_ketqua
```

```python
#Match events of insiders. df2 is the dataframe contains only events of affiliated persons
df2 = df[df['TypeName'] == 'GD CD noi bo']
df2 = df2[df2['ReportDate']<'2021-01-01']

#Get register events first, rename ReportDate as RegisteredDate, create df2_dangky & df2_ketqua 
df2_dangky = df2[df2['StatusName'] == 'Dang ky']
df2_dangky.rename(columns = {"ReportDate": "RegisterDate"}, inplace=True)
df2_dangky = df2_dangky[['StockCode','RegisterDate','DTTHCD','PositionCD','DateStartExpected','DateEndExpected']]

df2_ketqua = df2[df2['StatusName']=='Ket qua']
df2_ketqua.rename(columns={"ReportDate" : "ResultReportDate"}, inplace=True)
df2_dangky
```

```python
#Match df2_ketqua with df2_dangky using 'StockCode','DTTHCD','PositionCD','DateStartExpected','DateEndExpected'
df2_ketqua = df2_ketqua.merge(df2_dangky, left_on=['StockCode','DTTHCD','PositionCD','DateStartExpected','DateEndExpected'], right_on=['StockCode','DTTHCD','PositionCD','DateStartExpected','DateEndExpected'])
df2_ketqua
```

```python
df2_ketqua = df2_ketqua.reindex(columns=['StockCode','RegisterDate','ResultReportDate','DateStartExpected','DateEndExpected','DateActionFrom','DateActionTo','TypeName','RegisterVolumeBefore','RegisterVolumeBeforePercent','RegisterBuyVolume','RegisterSellVolume','RegisterVolumeAfter','RegisterVolumeAfterPercent','VolumeBefore','VolumeBeforePercent','BuyVolume','SellVolume','VolumeAfter','VolumeAfterPercent','Title_Event','PositionCD','DTTHCD','DTTHLQ','DTLQLQ','NVTH','ExtraPositionNLQ','ExtraPositionNLQEx','RelationShipType'])
df2_ketqua = df2_ketqua.dropna(subset=['RegisterDate','ResultReportDate','DateStartExpected','DateEndExpected','DateActionFrom','DateActionTo'])
df2_ketqua
```

```python
df1_ketqua['Diff_days'] = df1_ketqua['DateStartExpected'] - df1_ketqua['RegisterDate']
df1_ketqua['Diff_days'] = df1_ketqua['Diff_days'].dt.days
df1_ketqua = df1_ketqua[(df1_ketqua['Diff_days']>=3)&(df1_ketqua['Diff_days']<15)]
df1_ketqua
```

```python
df2_ketqua['Diff_days'] = df2_ketqua['DateStartExpected'] - df2_ketqua['RegisterDate']
df2_ketqua['Diff_days'] = df2_ketqua['Diff_days'].dt.days
df2_ketqua = df2_ketqua[(df2_ketqua['Diff_days']>=3)&(df2_ketqua['Diff_days']<15)]
df2_ketqua
```
